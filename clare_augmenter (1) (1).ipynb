```python
import warnings

# Mengabaikan semua peringatan
warnings.filterwarnings("ignore")
```

```python
import textattack
from textattack.augmentation import CLAREAugmenter

# Membuat CLAREAugmenter
clare_augmenter = CLAREAugmenter(pct_words_to_swap=0.1, transformations_per_example=1)
```

```python
# Contoh teks asli
original_text1 = "start each day with positive idea and make your day"

# Melakukan augmentasi
augmented_texts = clare_augmenter.augment(original_text1)
print(augmented_texts)
```

```python
path = 'C:/Users/USER/Documents/NLP/malay/augment_en.csv'
```

```python
df = pd.read_csv(path)
df
```

```python
def augment_text(text):
    augmented_texts = clare_augmenter.augment(text)
    # Mengembalikan teks augmentasi pertama
    return augmented_texts[0] if augmented_texts else text
```

```python
split = 5
df = df.iloc[:split]
```

```python
# Aplikasikan augmentasi pada kolom
df['clare_augment'] = df['translate_en'].apply(augment_text)
```
